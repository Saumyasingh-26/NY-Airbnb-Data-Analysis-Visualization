```python
data.drop(['id','host_id','host_name'], axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
#Checking for duplicates

data.duplicated().sum()
```

```python
data.fillna({"reviews_per_month":0, "name":0, "last_review":0}, inplace=True)
data.isnull().sum()
```

```python
print(data['neighbourhood_group'].nunique())
print(data['neighbourhood_group'].unique())
print(data['room_type'].nunique())
print(data['room_type'].unique())
```

```python
sns.histplot(data["price"],kde=True, bins=50, )
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
data['room_type'].value_counts().plot.barh().set_title("Room types frequency")
```

```python
data['neighbourhood_group'].value_counts().plot.barh().set_title("Neighbourhood frequency")
```

```python
data.groupby("neighbourhood_group").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(8,8))
plt.title('Pie graphic of neighbourhood')
```

```python
data.groupby('neighbourhood_group').agg({'price':'mean'}).plot.bar(alpha=1,color='black').set_title("Price mean of neighbourhood")
```

```python
data.groupby('neighbourhood_group').agg({'price':'mean','availability_365':'mean'}).plot.bar(alpha=0.7).set_title("Neighbourhood price and availability mean")
plt.legend(['Price','Availability'])
```

```python
df_1 = data.groupby(by=['neighbourhood_group', 'room_type']).agg({'price':'mean'})
df_1.reset_index(inplace=True)
print(df_1)
```

```python
sns.barplot(x=df_1['neighbourhood_group'],y=df_1['price'],hue=df_1['room_type'])
plt.title("Neighbourhood - Price - Room type")
```

```python
sns.scatterplot( x= data['price'], y=data['availability_365'])
```

```python
neighborhood_price_average= data.groupby('neighbourhood_group').agg({'price':'mean'})
sns.scatterplot( x= data['price'], y=data['minimum_nights'])
```

```python
sns.jointplot(x="price", y=data["neighbourhood_group"], data=data)
```

```python
sns.catplot(x="neighbourhood_group", y="price", data=data)
plt.title("Neighbourhood - Price")
```

```python
sns.catplot(data=data, x="neighbourhood_group", y="price", kind= 'strip' ,hue="room_type")
plt.title("Average price by neighborhood and room type.")
```

```python
sns.catplot(x="neighbourhood_group", y="price", hue="room_type", kind="point",data=data)
plt.title("Average price by neighborhood and room type.")
```
